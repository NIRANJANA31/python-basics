```python
#MATHS OPERATIONS ON DATA TYPE ARRAY
import numpy as np

#First Array
arr1 = np.array([[4,7],[2,6]],
                dtype = np.float64)
#second Arraky
arr2 = np.array([[3,6],[2,8]],
                dtype = np.float64)
#Addition of two Array
sum=np.add(arr1,arr2)
print(sum)
#Squares root of Array
Sqrt = np.sqrt(arr1)
print(Sqrt)

#Transpose of Array
Trans_arr = arr1.T
print(Trans_arr)
```

```python
# MATRIX OPERATIONS IN NUMPY
import numpy as np

x = np.array([[1, 2], [4, 5]])
y = np.array([[7, 8], [9, 10]])

print("Addition:\n", np.add(x, y))
print("Subtraction:\n", np.subtract(x, y))
print("Division:\n", np.divide(x, y))
```

```python
import numpy as np

x = np.array([[1, 2], [4, 5]])
y = np.array([[7, 8], [9, 10]])

print("Element-wise multiplication:\n", np.multiply(x, y))
print("Matrix multiplication:\n", np.dot(x, y))
```

```python
import numpy as np

x = np.array([[1, 2], [4, 5]])
y = np.array([[7, 8], [9, 10]])

print("Square root:\n", np.sqrt(x))
print("Sum of all elements:", np.sum(y))

print("Column-wise sum:", np.sum(y, axis=0))
print("Row-wise sum:", np.sum(y, axis=1))
print("Transpose:\n", x.T)
```

```python
#Task : Print Checkboard pattern using numpy and  matrix concept
import numpy as np

n = 8  # size of checkerboard (8x8)

checkerboard = np.zeros((n, n), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1

print(checkerboard)
```
